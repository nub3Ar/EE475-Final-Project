#### Introduction:
Our project is about building a handwritten digit sequence recognition applet using three different convolutional neural network architectures and the MNIST dataset. The MNIST dataset is a benchmark for image processing, and certain architectures have shown accuracy performance surpassing 99%. However, it is not very practical in real life since it is constrained to single digit recognition. Thus, we decided to utilize existing literature and build a successful end-to-end pipeline for recognizing not a single, but a sequence of, handwritten digits in this project. This would have applications in any fields that require recognition of numeric sequences, such as contact collection, retail banking, and education.

#### Image Pre-processing
Since we aim to recognize a sequence of handwritten digits, the first step is to pre-process our image so that we split up the image into an array of matrices, each of which represents a single digit. Given that we trained our model using the MNIST dataset, we need to make sure that our image is also black-and-white and each digit is of the same dimension (28*28). We split up each digit by going through our matrix using for loop, and for the first black pixel we see in a column, we mark that as the left of a digit, and for the first column that’s entirely white, we mark it as the right of that digit. Then, we add appropriate amount of white pixel padding to make sure both horizontal and vertical dimensions are divisible by 28. In the end, we will pass the n*28*28*1 array to our trained models for prediction. 

#### Model Architecture:
We have explored different state-of-the-art architectures for MNIST digit recognition and compared their performance. A recent extensive survey of MNIST-classification networks was done by Baldominos et al., 2019 (ref.) Here, we implemented and compared three neural network architectures. Network denoted as CNN is a convolutional network with two convolutional layers with MaxPool layer and with ReLU activation function followed by a dense layer and softmax activation for the classification task. CNN_Drop is based on the CNN and adds a Dropout layer before the final softmax activation function. CNN with Batch Normalization layers is also based on the standard CNN architecture and adds two batch normalization layers before MaxPool functions.

#### Shortcomings and Future Work
Admittedly, there are some shortcomings in our current work. The first is that the input image is limited in its type (we only accept .jpg for now), and we don’t support dynamic cropping/scaling of the image. Another issue is that our current model architecture is not robust enough to handle all input sizes (limited to 28*28 image size). In the future, we can work on solving the size and type limitations, which would enable a wider and more flexible usage of our applet. Another area of improvement is our limited and potentially non-representative training data.  Since we trained our models with the MNIST dataset, but used the models to predict our own handwritten digits, we can see a noticeable accuracy drop for such previously unseen data. One potential solution is to train on native data instead of MNIST (use amazon mule to generate enough training data and retrain all three networks). 

```python
import tensorflow as tf
import keras
from keras import layers
import numpy as np
import matplotlib.pyplot as plt
from utils.image_processor import transformation
from classifier.models import CNN
from classifier.Main import CNN_prediction
from utils.util_function import show_images, plot_losses, plot_accuracy, process_data
```

#### Load data

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, y_train, x_test, y_test = process_data(x_train, y_train, x_test, y_test)
```

```python
INPUT_SHAPE = (28,28,1)
NUM_CLASSES = 10

class CNN(keras.Model):

    def __init__(self):
        super(CNN, self).__init__()
        self.model = keras.models.Sequential()
        self.model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), strides=1, padding='Same', activation='relu',
                                      input_shape = INPUT_SHAPE) )
        self.model.add(keras.layers.MaxPool2D(pool_size=(2, 2), padding='same'))
        self.model.add(keras.layers.Conv2D(filters=64,kernel_size=(3, 3),strides=1,padding='same', activation='relu'))
        self.model.add(keras.layers.MaxPool2D(pool_size=(2, 2),padding='same'))
        self.model.add(keras.layers.Flatten())
        self.model.add(keras.layers.Dense(units=NUM_CLASSES, activation='sigmoid'))

    def call(self, inputs):
        return self.model(inputs)

class CNN_Drop(keras.Model):

    def __init__(self):
        super(CNN_Drop, self).__init__()
        self.model = keras.models.Sequential()
        self.model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), strides=1, padding='Same', activation='relu',
                                      input_shape = INPUT_SHAPE) )
        self.model.add(keras.layers.MaxPool2D(pool_size=(2, 2), padding='same'))
        self.model.add(keras.layers.Conv2D(filters=64,kernel_size=(3, 3),strides=1,padding='same', activation='relu'))
        self.model.add(keras.layers.MaxPool2D(pool_size=(2, 2),padding='same'))
        self.model.add(keras.layers.Flatten())
        self.model.add(keras.layers.Dropout(0.5))
        self.model.add(keras.layers.Dense(units=NUM_CLASSES, activation='sigmoid'))

    def call(self, inputs):
        return self.model(inputs)

class CNN_BN(keras.Model):

    def __init__(self):
        super(CNN_BN, self).__init__()
        self.model = keras.models.Sequential()
        self.model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), strides=1, padding='Same', input_shape = INPUT_SHAPE) )
        self.model.add(keras.layers.BatchNormalization())
        self.model.add(keras.layers.ReLU())
        self.model.add(keras.layers.MaxPool2D(pool_size=(2, 2), padding='same'))
        
        self.model.add(keras.layers.Conv2D(filters=64,kernel_size=(3, 3),strides=1,padding='same'))
        self.model.add(keras.layers.BatchNormalization())
        self.model.add(keras.layers.ReLU())
        self.model.add(keras.layers.MaxPool2D(pool_size=(2, 2),padding='same'))
        
        self.model.add(keras.layers.Flatten())
        self.model.add(keras.layers.Dense(units=NUM_CLASSES, activation='sigmoid'))

    def call(self, inputs):
        return self.model(inputs)
```

#### Training

```python
epochs = 20
batch_size = 128

# regular CNN
cnn = CNN()
cnn.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

CNN_history = cnn.fit(x_train[:10000], y_train[:10000], batch_size=batch_size, epochs=epochs, validation_split=0.1, verbose=0)
score = cnn.evaluate(x_test, y_test, verbose=0)
print("CNN \n", "Test loss:", score[0], "\n Test accuracy:", score[1])
cnn.save_weights("ckpt_CNN")

# CNN with dropout
cnn_drop = CNN_Drop()
cnn_drop.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

CNN_drop_history = cnn_drop.fit(x_train[:10000], y_train[:10000], batch_size=batch_size, epochs=epochs, validation_split=0.1, verbose=0)
score = cnn_drop.evaluate(x_test, y_test, verbose=0)
print("CNN with Dropout \n","Test loss:", score[0], "\n Test accuracy:", score[1])
cnn_drop.save_weights("ckpt_CNN_Drop")

# CNN with batch normalization
cnn_bn = CNN_BN()
cnn_bn.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

CNN_BN_history = cnn_bn.fit(x_train[:10000], y_train[:10000], batch_size=batch_size, epochs=epochs, validation_split=0.1, verbose=0)
score = cnn_bn.evaluate(x_test, y_test, verbose=0)
print("CNN with BN \n", "Test loss:", score[0], "\n Test accuracy:", score[1])
cnn_bn.save_weights("ckpt_CNN_BN")
```

#### Results

```python
plot_losses([CNN_history, CNN_drop_history, CNN_BN_history ], ['CNN', 'CNN Dropout', 'CNN BatchNorm'])
plot_accuracy([CNN_history, CNN_drop_history, CNN_BN_history], ['CNN', 'CNN Dropout', 'CNN BatchNorm'])
```
